```python
import pandas as pd
import matplotlib.pyplot as plt


url = 'https://raw.githubusercontent.com/jre29/lecture1/refs/heads/main/Salary%20Data%20(2).csv'
data = pd.read_csv(url)


data.head()
plt.figure(figsize=(10, 6))
plt.scatter(data['Years of Experience'], data['Salary'], color='blue')
plt.xlabel('Lata doświadczenia')
plt.ylabel('Wynagrodzenie')
plt.title('Zależność wynagrodzenia od lat doświadczenia')
plt.grid()
plt.show()
```
